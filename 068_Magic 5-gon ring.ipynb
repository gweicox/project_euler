<p>By concatenating each group it is possible to form 9-digit strings; the maximum string for a 3-gon ring is 432621513.</p>

<p>Using the numbers 1 to 10, and depending on arrangements, it is possible to form 16- and 17-digit strings.<br>
   What is the maximum <b>16-digit</b> string for a "magic" 5-gon ring?</p>

<div class="center" style="transform: scale(0.7);">



</div>

```python
%%time


from itertools import permutations


def algos():
    n_max, a = 0, 6

    for b, c, d, e in permutations((7, 8, 9, 10)):
        for v, w, x, y, z in permutations((1, 2, 3, 4, 5)):
            if (a + v + w) == (b + w + x) == (c + x + y) == (d + y + z) == (e + z + v):
                n = int(''.join((map(str, [a, v, w, b, w, x, c, x, y, d, y, z, e, z, v]))))
                n_max = max(n_max, n)

    return n_max


algos()

# 6531031914842725
# 2024.02.06
# Wall time: 0 ns
```
