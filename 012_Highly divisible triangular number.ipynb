### 12. Highly divisible triangular number

link: https://projecteuler.net/problem=12

The sequence of triangle numbers is generated by adding the natural numbers. <br>
So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 **1**: 1  
 **3**: 1, 3  
 **6**: 1, 2, 3, 6  
**10**: 1, 2, 5, 10  
**15**: 1, 3, 5, 15  
**21**: 1, 3, 7, 21  
**28**: 1, 2, 4, 7, 14, 28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
from math import prod

def prime_factors_powers(n: int) -> dict:
    res, d = {}, 2
    while d * d <= n:
        if n % d == 0: res[d] = res.get(d, 0) + 1; n //= d
        else: d += 1 
    if n > 1: res[n] = res.get(n, 0) + 1
    return res

def divisors_number(n: int) -> int:
    return prod(x + 1 for x in prime_factors_powers(n).values())
```

writing the number as a product of prime factors:
$$ n = p^aq^br^c... $$
then the number of divisors: 
$$ d(n) = (a+1)(b+1)(c+1)... $$

```python
%%time

k, num = 7, 28 #sum(range(k + 1))

while True:
    d = divisors_number(num)
    
    if d > 500:
        print(f'k = {k}, num = {num}, d = {d}\n')
        break
    else:
        k += 1
        num += k
        
# 2022.07.02
```
