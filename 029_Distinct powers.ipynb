### 29. Distinct powers

link: https://projecteuler.net/problem=29

Consider all integer combinations of a<sup>b</sup> for 2 ≤ _a_ ≤ 5 and 2 ≤ _b_ ≤ 5:

> 2^2=4, 2^3=8, 2^4=16, 2^5=32  
> 3^2=9, 3^3=27, 3^4=81, 3^5=243  
> 4^2=16, 4^3=64, 4^4=256, 4^5=1024  
> 5^2=25, 5^3=125, 5^4=625, 5^5=3125  

If they are then placed in numerical order, with any repeats removed, we get the following sequence of 15 distinct terms:

4, 8, 9, 16, 25, 27, 32, 64, 81, 125, 243, 256, 625, 1024, 3125

How many distinct terms are in the sequence generated by a<sup>b</sup> for 2 ≤ _a_ ≤ 100 and 2 ≤ _b_ ≤ 100?


```python
%%time


len({a ** b for a in range(2, 101) for b in range(2, 101)})


# 9183
# 2022.07.19
# Wall time: 6 ms
```
